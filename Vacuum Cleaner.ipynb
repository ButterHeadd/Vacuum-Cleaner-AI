```python
import random
```

```python
class VacuumCleaner:
    def __init__(cleaner):
        #Generates a dictionary that acts as the environment, also randomizes the cleaners location
        cleaner.location = random.choice([1, 2, 3, 4, 5, 6, 7, 8, 9])
        print(f"Started in {cleaner.location}")
        cleaner.environment = {
        1 : "Dirt",
        2 : "Dirt",
        3 : "Dirt",
        4 : "Clean",
        5 : "Clean",
        6 : "Clean",
        7 : "Clean",
        8 : "Clean",
        9 : "Clean",
        }
        
    def percept(cleaner):
        #returns the location and if its clean or not
        return cleaner.location, cleaner.environment[cleaner.location]

    def perceptright(cleaner):
        #checks to the right of the cleaner, to see if its clean or dirty in that space
        if cleaner.location == 9:
            cleaner.locationright = cleaner.location
        else:
            cleaner.locationright = cleaner.location + 1
        return cleaner.environment[cleaner.locationright]

    def perceptleft(cleaner):
        #checks to the left of the cleaner, to see if its clean or dirty in that space
        if cleaner.location == 1:
            cleaner.locationleft = cleaner.location
        else:
            cleaner.locationleft = cleaner.location - 1
        return cleaner.environment[cleaner.locationleft]

    def perceptclosest(cleaner):
        #Checks all spaces to the right or left of it, and determines which dirt space is furthest away.
        left = 0 
        right = 0
        steps2 = 0
        locationleft = cleaner.location - 1
        locationright = cleaner.location + 1
        if cleaner.location != 1 and locationleft >= 1:
            while locationleft >= 1:
                steps2+=1
                locationleft = cleaner.location - steps2
                if locationleft > 0 and cleaner.environment[locationleft] == "Dirty":
                    left = steps2
                    break
                elif locationleft == 1 and cleaner.environment[locationleft] == "Clean":
                    left = 9
                    break
        steps2 = 0
        if cleaner.location != 9 and locationright <= 9:
            while locationright <= 9:
                steps2+=1
                locationright = cleaner.location + steps2
                if locationright < 10 and cleaner.environment[locationright] == "Dirty":
                    right = steps2
                    break
                elif locationright == 9 and cleaner.environment[locationright] == "Clean":
                    right = 9
                    break
        if left > right:
            direction = 1 #go right
            return direction
        elif left == right:
            direction = random.choice([1, 2]) #random
            return direction
        elif right > left:
            direction = 2 #go left
            return direction
            
    def act(cleaner, perception):
        #moves the cleaner randomly, cleaning if there's dirt on the space its at.
        location, status = perception
        if status == "Dirt":
            print(f"Vacuum cleaning dirt at {location}.")
            cleaner.environment[location] = "Clean"
        else:
            if cleaner.location == 1:
                cleaner.location = 2
            elif cleaner.location == 9:
                cleaner.location = 8
            else:
                direction = random.choice([1, 2])
                if direction == 1:
                    cleaner.location = cleaner.location + 1
                elif direction == 2: 
                    cleaner.location = cleaner.location - 1
            print(f"Room is clean. Moving to {cleaner.location}.")

    def smartAct(cleaner, perception, statusright, statusleft):
        #checks right or left before moving, if both sides are clean then moves randomly
        location, status = perception
        statusright = statusright
        statusleft = statusleft
        if status == "Dirt":
            print(f"Vacuum cleaning dirt at {location}.")
            cleaner.environment[location] = "Clean"
        else:
            if cleaner.location == 1:
                cleaner.location = 2
            elif cleaner.location == 9:
                cleaner.location = 8
            else:
                if statusright == "Dirt":
                    print("Dirt to the right.")
                    cleaner.location = cleaner.location + 1
                elif statusleft == "Dirt":
                    print("Dirt to the left.")
                    cleaner.location = cleaner.location - 1
                else:
                    direction = random.choice([1, 2])
                    if direction == 1:
                        cleaner.location = cleaner.location + 1
                    elif direction == 2:
                        cleaner.location = cleaner.location - 1
            print(f"Room is clean. Moving to {location}.")

    def smartestAct(cleaner, perception, direction):
        #checks closest dirt space then moves towards that
        location, status = perception
        direction = direction
        nextlocation = 0
        if status == "Dirt":
            print(f"Vacuum cleaning dirt at {location}.")
            cleaner.environment[location] = "Clean"
            cleaner.fuelsteps += 1
        else:
            if cleaner.location == 1:
                cleaner.location = 2
                nextlocation = cleaner.location
            elif cleaner.location == 9:
                cleaner.location = 8
                nextlocation = cleaner.location
            else:
                if direction == 1:
                    cleaner.location = cleaner.location + 1
                    nextlocation = cleaner.location
                elif direction == 2:
                    cleaner.location = cleaner.location - 1
                    nextlocation = cleaner.location
            print(f"Room is clean. Moving to {nextlocation}.")

    def fuelCost(cleaner):
        #Determines how much electricity the cleaner takes
        #vacuums take 500 to 1500 watts an hour, on average. so i divided 1000 by 6, 
        #to simulate each clean taking 10 mins on average, so its 167 (166.6 recurring) per 10 mins.
        fuel = 167 * cleaner.fuelsteps
        print(f"Fuel at: {fuel}.")
        
    def goalState(cleaner):
        #What the cleaner is working towards
        if all(status == "Clean" for status in cleaner.environment.values()):
            print("All rooms are clean.")
        return all(status == "Clean" for status in cleaner.environment.values())

    def run(cleaner):
        #basically the main function, runs the code.
        typeofai = input("1. Normal AI \n2. Smart AI \n3. Smartest AI\n ")
        typeofai = int(typeofai)
        steps = 0
        cleaner.fuelsteps = 0
        if typeofai == 1:
            while not cleaner.goalState():
                perception = cleaner.percept()
                cleaner.act(perception)
                steps += 1
                print(f"Step {steps}: \n Environment state: {cleaner.environment}\n")
        elif typeofai == 2:
            while not cleaner.goalState():
                perception = cleaner.percept()
                statusright = cleaner.perceptright()
                statusleft = cleaner.perceptleft()
                cleaner.smartAct(perception, statusright, statusleft)
                steps += 1
                print(f"Step {steps}: \n Environment state: {cleaner.environment}\n")
        elif typeofai == 3:
            while not cleaner.goalState():
                perception = cleaner.percept()
                direction = cleaner.perceptclosest()
                cleaner.smartestAct(perception, direction) 
                cleaner.fuelCost()
                steps += 1
                print(f"Step {steps}: \n Environment State: {cleaner.environment}\n")

agent = VacuumCleaner()
agent.run()
```
